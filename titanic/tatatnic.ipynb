```python
from sklearn.model_selection import train_test_split
X_train,x_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=2024)
```

```python
from sklearn.ensemble import RandomForestClassifier,AdaBoostClassifier
from sklearn.naive_bayes import MultinomialNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
models=[RandomForestClassifier(n_estimators=120),AdaBoostClassifier(n_estimators=70),MultinomialNB(),KNeighborsClassifier(n_neighbors=3),SVC(),LogisticRegression()]
for x in models:
    print(x.fit(X_train,y_train))
    print(x.score(x_test,y_test))
```

```python
model=RandomForestClassifier(n_estimators=120)
```

```python
model.fit(X_train,y_train)
```

```python
model.score(x_test,y_test)
```

```python
y_pred=model.predict(x_test)
```

```python
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
print(classification_report(y_test,y_pred))
```

```python
sns.heatmap(data=confusion_matrix(y_test,y_pred),annot=True,cmap='viridis')
```
